# Task 5: Reasons for Investment

Objective: Analyze and summarize reasons
for investment choices.
Steps:

1.**Explore Reasons Column**: Investigate

2.**Summarize Reasons**: Identify common

```python
# Task 5: Reasons for Investment Choices
import matplotlib.pyplot as plt
import numpy as np

# Data Preparation
reasons = ["Risk Free", "Assured Returns", "Interest Rates"]
categories = ["Equity", "Mutual Funds", "Bonds", "Fixed Deposits"]

data = [
    [10, 5, 2],  # Equity
    [8, 6, 3],   # Mutual Funds
    [7, 4, 1],   # Bonds
    [12, 7, 2]   # Fixed Deposits
]

# Convert to NumPy array for better manipulation
data = np.array(data)
bar_width = 0.2

# Positions of bars on X-axis
x = np.arange(len(reasons))

# Define new colors for better clarity
colors = ["orange", "blue", "green", "brown"]

plt.figure(figsize=(9, 5))

# Plot each category as separate bars
for i in range(len(categories)):
    plt.bar(x + i * bar_width, data[i], width=bar_width, label=categories[i], color=colors[i])

# Title and labels
plt.title("Reasons for Investment Choices", fontsize=14, fontweight="bold")
plt.xlabel("Investment Reasons", fontsize=12)
plt.ylabel("Count", fontsize=12)

# X-axis labels formatting
plt.xticks(x + bar_width * 1.5, reasons, rotation=30, ha="right", fontsize=10)

# Display legend
plt.legend(title="Investment Type")
plt.tight_layout()
plt.show()
```

# Task 7: Common Information Sources

Objective: Analyze common sources participants
rely on for investment information.

Steps:

1.**Explore Information Sources Column**: Review
the column where participants indicated their
sources of investment information.

2.**Identify Common Sources**: Analyze the data to
identify and summarize the most common
sources participants rely on.


```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns

print("\n--- Task 10: Correlation Analysis ---")

# Load the dataset
file_path = "/content/Data_set 2 - Copy.csv"  # Update the path if needed
df = pd.read_csv(file_path)

# Step 1: Check available column names
print("Available Columns in Dataset:")
print(df.columns)

# Step 2: Identify 'Expect' column (case-insensitive)
expected_returns_col = None
for col in df.columns:
    if "expect" in col.lower():
        expected_returns_col = col
        break

if expected_returns_col:
    print(f"Using column: '{expected_returns_col}' for Expected Returns conversion.")
else:
    print("Error: 'Expect' column not found. Check dataset column names.")
    expected_returns_col = None  # Don't assume a default if not found

# Step 3: Function to convert ranges to numeric values
def convert_range_to_numeric(value):
    if isinstance(value, str):
        numbers = [float(s) for s in re.findall(r'\d+', value)]  # Extract numbers
        if len(numbers) == 2:
            return np.mean(numbers)  # Average for ranges
        elif len(numbers) == 1:
            return numbers[0]  # Single value case
    return np.nan  # Handle missing values

# Step 4: Apply conversion to "Expect" if the column exists
if expected_returns_col:
    df["Expect"] = df[expected_returns_col].apply(convert_range_to_numeric)
    df["Expect"] = df["Expect"].fillna(df["Expect"].median())  # Fix: Direct assignment

    print("\nFirst 10 values in 'Expect' column AFTER conversion:")
    print(df["Expect"].head(10))
else:
    print("Skipping 'Expect' conversion due to missing column.")

# Step 5: Convert "Duration" to numeric values
if "Duration" in df.columns:
    df["Duration"] = df["Duration"].apply(convert_range_to_numeric)
    df["Duration"] = df["Duration"].fillna(df["Duration"].median())  # Fix: Direct assignment

# Step 6: Define numerical columns for correlation
selected_columns = ["age", "Duration", "Expect"]
valid_columns = [col for col in selected_columns if col in df.columns]

if len(valid_columns) == len(selected_columns):
    # Check for missing values
    print("\nMissing Values in Selected Columns:")
    print(df[selected_columns].isnull().sum())

    # Compute correlation matrix
    correlation_matrix = df[selected_columns].corr()
    print("\nCorrelation Matrix:")
    print(correlation_matrix)

    # Heatmap visualization
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.show()

    # Scatter plot matrix
    sns.pairplot(df[selected_columns].dropna())  # Ensure no NaN values
    plt.show()

else:
    print("\nSome selected columns are missing in the dataset. Check column names and update the list.")

print("\nAnalysis completed successfully!")

```
